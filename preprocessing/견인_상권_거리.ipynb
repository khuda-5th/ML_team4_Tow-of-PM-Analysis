```python
import pandas as pd

# 데이터 불러오기
data = pd.read_csv('/버스_지하철_거리계산.csv', encoding='cp949')  # 파일 경로를 적절히 수정해주세요.

# 필요한 열만 선택
filtered_data = data[['주소', '횟수', '위도', '경도']]

filtered_data
```

```python
# 데이터 불러오기
data2 = pd.read_csv('/유동인구_영역.csv', encoding='cp949')  # 파일 경로를 적절히 수정해주세요.
data2
```

```python
from math import radians, sin, cos, sqrt, atan2
import numpy as np
import pandas as pd

# 거리 계산 함수
def haversine(lat1, lon1, lat2, lon2):
    """
    두 위도 경도 좌표 사이의 거리를 계산하는 함수
    """
    # 지구의 반경 (km)
    R = 6371.0

    # 라디안 변환
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    # 위도, 경도 차이 계산
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    # 하버사인 공식을 사용하여 거리 계산
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# 거리 정보를 저장할 리스트 초기화
distances = np.zeros((len(filtered_data), len(data2)))

# filtered_data의 각 행에 대해 data2의 각 행과의 거리 계산
for i, row in filtered_data.iterrows():
    for j, row2 in data2.iterrows():
        distances[i, j] = haversine(row['위도'], row['경도'], row2['위도'], row2['경도'])

# 거리 정보를 포함한 데이터프레임 생성
distance_df = pd.DataFrame(distances, columns=[f'distance_{index}' for index in range(len(data2))])

# 결과 데이터프레임 생성
result_df = pd.concat([filtered_data.reset_index(drop=True), distance_df], axis=1)

# 결과 데이터프레임의 크기 확인
print("결과 데이터프레임의 크기:", result_df.shape)
```

```python
result_df.to_csv('result_df.csv', index=False)
```
