```python
import pandas as pd
```

```python
df_origin = pd.read_csv("/서울시 상권분석서비스(길단위인구-상권배후지).csv", encoding='cp949')
```

```python
df_origin['기준_년분기_코드'] = df_origin['기준_년분기_코드'].astype(str)
df = df_origin[df_origin['기준_년분기_코드'].str.startswith('2023')]
df['기준_년분기_코드'] = df['기준_년분기_코드'].astype(int)
```

```python
df
```

```python
df['연간_유동인구_수'] = df.groupby('상권배후지_코드_명')['총_유동인구_수'].transform('sum')

print(df)
```

```python
df = df.drop_duplicates(subset='상권배후지_코드_명', keep='first')
print(df)
```

```python
print(df.columns)
```

```python
df = df[['상권배후지_코드', '상권배후지_코드_명', '연간_유동인구_수']]
df = df.reset_index(drop=True)
```

```python
df
```

```python
df2 = pd.read_csv("/서울시 상권분석서비스(영역-상권).csv", encoding='cp949')
```

```python
df2
```

```python
# '행정동_코드_명' 칼럼만 추출하여 df2를 새로운 DataFrame으로 생성
df2_subset = df2[['상권_코드_명', '행정동_코드_명', '엑스좌표_값', '와이좌표_값']]

# '상권배후지_코드_명'과 '상권_코드_명'이 일치하는 행만 병합
merged_df = pd.merge(df, df2_subset, left_on='상권배후지_코드_명', right_on='상권_코드_명', how='left')
merged_df.drop('상권배후지_코드_명', axis=1, inplace=True)

print(merged_df)
```

```python
from pyproj import Proj, transform

# EPSG 5181 좌표계에 대한 프로젝션 정의
in_proj = Proj(init='epsg:5181')
out_proj = Proj(init='epsg:4326')  # WGS 84 좌표계

# merged_df에서 엑스좌표_값과 와이좌표_값 열을 추출하여 위경도로 변환 후 새로운 열에 추가
merged_df['경도'], merged_df['위도'] = transform(in_proj, out_proj, merged_df['엑스좌표_값'].values, merged_df['와이좌표_값'].values)

merged_df.drop('엑스좌표_값', axis=1, inplace=True)
merged_df.drop('와이좌표_값', axis=1, inplace=True)
merged_df.drop('상권배후지_코드', axis=1, inplace=True)


merged_df = merged_df.sort_values(by='연간_유동인구_수', ascending=False)
merged_df = merged_df.reset_index(drop=True)

print(merged_df)
```

```python
merged_df.to_csv('유동인구_영역.csv', encoding = 'cp949')
```
