```python
import pandas as pd

# CSV 파일 경로 지정
file_path = '/content/201_DT_201003_A010006_20240228235534.csv'

# 특정 행을 제외하고 데이터 불러오기
rows_to_skip = [29832]  # 제외할 행의 번호를 리스트로 지정
df = pd.read_csv(file_path, skiprows=rows_to_skip)

# 불러온 데이터 확인
print(df)
```

```python
# '동별'로 그룹화하고 총 그룹의 갯수 확인
total_dong_groups = len(df.groupby('동별'))

# 결과 출력
print("총 그룹화된 동의 갯수:", total_dong_groups)

# 여기엔 생각해보니까 .. 구랑 이상한 합계가 들어가는구나 지워야겠네
```

```python
import re

# 각 행에 대해 '~동' 추출하는 함수
def extract_dong_name(address):
    match = re.search(r'(\w+동)', address)
    return match.group(1) if match else None

# df_tow_location 데이터프레임에 새로운 열 추가
df_tow_location['동'] = df_tow_location['주소'].apply(extract_dong_name)

# 결과 확인
print(df_tow_location[['주소', '동']])
```

```python
import requests

# 발급받은 카카오맵 API 키
kakao_api_key = '838ed4e900821a97b1a6244b29ff14e1'

def get_dong_from_address(address):
    try:
        # 주소로부터 행정동 정보를 조회하는 API 요청
        url = f'https://dapi.kakao.com/v2/local/search/address.json'
        headers = {'Authorization': f'KakaoAK {kakao_api_key}'}
        params = {'query': address}
        response = requests.get(url, headers=headers, params=params)
        data = response.json()

        # API 응답에서 행정동 정보 추출
        dong = data['documents'][0]['address']['region_3depth_h_name']

        return dong
    except Exception as e:
        print(f'Error fetching data for address {address}: {str(e)}')
        return None

# 주소에 대해 행정동 정보 가져오기
df_tow_location['행정동'] = df_tow_location['주소'].apply(get_dong_from_address)

# 결과 확인
print(df_tow_location[['주소', '행정동']])
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y
    except:
        return [0,0]
```

```python
latitude = []
longitude =[]

for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```
